# Keyword Search 

```python
#Parameters
API = 'authorizedtiles'
END_POINT = BASE_URL + API
search_text = "housing starts"
selected_facets = []

#Execute search
data = requests.post(url = END_POINT, params = {'api_key': API_KEY}, json = {"searchText": search_text,"selectedFacets": selected_facets})
search_result = data.json()

#Convert to pandas dataframe
data_tiles = search_result['dataTiles']
data_frame = pandas.DataFrame(data_tiles)
data_frame
```

# Selected top 100 tiles

```python
priority = 0
selected_data = []
for data_tile in data_tiles[0:100]:
    dt_code = data_tile['ID']
    priority = priority + 1
    selected_data.append({"dataTileCode":dt_code,"Priority": priority})
print(selected_data[0:10])
```

# Run Publication

```python
#Parameters
API = 'runpublication'
END_POINT = BASE_URL + API
file_name = 'NAC-1022 Publication with top 100 housing starts tiles '
download_format = {"format":"CSV","compression":"NONE","download_when_ready": False}
subscription_frequency = 'one-time'

#Execute API
data = requests.post(url = END_POINT, params = {'api_key': API_KEY}, json = {'selected_data': selected_data, 'file_name': file_name, 'download_format': download_format, 'subscription_frequency': subscription_frequency}) 
data = data.json()
publication_id = data['publication_id']
print(data)
```

# Check Publication Status

```python
# Check for publication completion
API = 'publicationstatus'
END_POINT = BASE_URL + API
data = requests.post(url = END_POINT, params = {'api_key': API_KEY}, json = {'publication_id': publication_id}) 
print(data.json())
```
